### Zistovanie, ci nasli sme naozaj ten povodny dataset

```python
df_train = pd.read_csv('data/heart.csv',delimiter=';')
df_train['oldpeak'] = df_train['oldpeak'].str.replace(',','.')
df_train = df_train.astype(float)
df_train.rename(columns={'target,':'target'}, inplace=True)

df_test = pd.read_csv('data/heart_test.csv', delimiter=";")
df_test = df_test.astype(float)
```

```python
df_test = df_test.loc[:, df_train.columns.tolist()]
```

```python
df = pd.concat([df_train, df_test], axis=0).reset_index(drop=True)
```
